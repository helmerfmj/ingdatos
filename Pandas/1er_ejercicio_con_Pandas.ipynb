```python
#Importamos la librería de Pandas y la denominamos como pd
import pandas as pd
```

```python
data = {
    'nombre': ['Ana', 'Juan', 'María'],
    'edad': [25, 30, 28],
    'ciudad': ['Madrid', 'Barcelona', 'Sevilla']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
print(df.describe())
print(df['edad'].mean())
print(df.groupby('ciudad').count())
```
